```python
import os
import os.path
import pandas as pd

datadir = 'publicdata'
file1name = 'namespop.csv'
file2name = 'birthrate.csv'
```

```python
path = os.path.join(datadir, file1name)
assert os.path.isfile(path)
namespop = pd.read_csv(path)
```

```python
path = os.path.join(datadir, file2name)
assert os.path.isfile(path)
birthrate = pd.read_csv(path, index_col=0)
```

```python
def near_year(year, years):
    for refyear in years:
        if refyear >= year:
            return refyear
    return years[-1]
```

```python
refyears = list(birthrate.index)
```

```python
namespop['birth_rate'] = namespop['year'].apply(lambda y: birthrate.loc[near_year(y, refyears)])
```

```python
namespop.iloc[30:40]
```

```python
namespop['births'] = namespop['population'] / 1000 * namespop['birth_rate']
```

```python
namespop
```

```python
namespop['births'] = namespop['births'].apply(int)
```

```python
names2 = namespop[['year', 'sex', 'name', 'count', 'births']]
```

```python
names2.to_csv(os.path.join(datadir, "namesbirths.csv"), index=False)
```
