# Markdown Practice Instructions

The objective of this part of the assignment is to give you practice with Markdown as a means of constructing documents within our Jupyter notebook framework.

Markdown takes basic text and adds a limited number of syntax and organizational annotations that, like LaTeX or HTML, allow a system to generate a rendering for a display or print version.  But at the same time, the annotations are less obtrusive and allow the text to **still be readable** even without rendering.  This allows exchange of documents without worrying about the software capabilities, and allows platforms like these notebooks, or even the discussion and other elements of Notebowl, our learning management system, to use and render text to allow clearer communication.
